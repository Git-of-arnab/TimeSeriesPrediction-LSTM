```python
import numpy as np
import pandas as pd
raw_alarm = pd.read_csv('/content/billa_netact_alarms.csv')
raw_alarm.head(3)
```

```python
raw_alarm[['Last Detected Date','Last Detected Time','Time Zone']] = raw_alarm['Last Time Detected'].str.split(" ", n=2, expand = True)
raw_alarm.drop(columns = ['Last Time Detected','First Time Detected'], inplace = True)
raw_alarm.head(3)
```

```python
raw_alarm['Last Detected'] = raw_alarm['Last Detected Date']+' '+raw_alarm['Last Detected Time']
raw_alarm['Last Detected'] = pd.to_datetime(raw_alarm['Last Detected'])
raw_alarm.head(2)
```

```python
raw_alarm['unit'] = bb_raw_alarm[0]
```

```python
xx_raw_alarm = raw_alarm['Alarm Name'].str.split("_",n=1,expand = True)
xx_raw_alarm.head()
```

```python
raw_alarm['Alarm Name'] = xx_raw_alarm[1]
raw_alarm.head()
```

```python
#Defining the train and test data
X = final_dataset.drop(columns = 'Attach Fail', axis =1)
Y = final_dataset['Attach Fail']
print('The shape of X features is: ',X.shape)
print('The shape of Y output is: ', Y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test,Y_train, Y_test = train_test_split(X,Y, test_size=0.10)

print('The shape of X features is: ',X.shape)
print('The shape of Y output is: ', Y.shape)
print('The shape of X_train features is: ',X_train.shape)
print('The shape of Y_train is: ',Y_train.shape)
print('The shape of X_test  is: ', X_test.shape)
print('The shape of Y_test is: ', Y_test.shape)
```

```python
print('The type of X features is: ',type(X))
print('The type of Y output is: ', type(Y))
print('The type of X_train features is: ',type(X_train))
print('The type of X_test  is: ', type(X_test))
print('The type of Y_train is: ',type(Y_train))
print('The type of Y_test is: ', type(Y_test))
```

```python
#converting our train data into numpy array as NN takes only numpy in tensorflow
x_training_data = np.array(X_train)

y_training_data = np.array(Y_train)
print(x_training_data.shape)

print(y_training_data.shape)
```

```python
x_training_data = np.expand_dims(x_training_data, axis = -1)
print(x_training_data.shape)
```

```python
#x_training_data = np.reshape(x_training_data, (x_training_data.shape[0],x_training_data.shape[1],1))
#print(x_training_data.shape)
```

```python
print(x_training_data.shape[0])
print(x_training_data.shape[1])
```

```python
#converting our test data into numpy array as NN takes only numpy in tensorflow
x_test_data = np.array(X_test)

y_test_data = np.array(Y_test)
print(x_test_data.shape)

print(y_test_data.shape)
```

```python
x_test_data = np.expand_dims(x_test_data, axis = -1)
print(x_test_data.shape)
```

```python
from tensorflow.keras.models import Sequential

from tensorflow.keras.layers import Dense

from tensorflow.keras.layers import LSTM

from tensorflow.keras.layers import Dropout
```

```python
from keras.api._v2.keras import activations
rnn = Sequential()
rnn.add(LSTM(units = 45, return_sequences = True, input_shape = (x_training_data.shape[1],1)))
#input should be 3D with shape [batch, timesteps, feature]
rnn.add(Dropout(0.2))

rnn.add(LSTM(units = 45, return_sequences = True))

rnn.add(Dropout(0.2))

rnn.add(LSTM(units = 30, return_sequences = True))

rnn.add(Dropout(0.2))

rnn.add(LSTM(units = 25))

rnn.add(Dropout(0.2))
rnn.add(Dense(units = 1, activation = 'sigmoid'))
```

```python
rnn.summary()
```

```python
import tensorflow as tf
rnn.compile(optimizer = 'adam', loss = tf.keras.losses.BinaryCrossentropy(from_logits=True),metrics=['accuracy'])
```

```python
rnn.fit(x_training_data, y_training_data, epochs = 150, batch_size = 32)
```

```python
# evaluate the model
_, train_acc = rnn.evaluate(x_training_data, y_training_data, verbose=0)
_, test_acc = rnn.evaluate(x_test_data, y_test_data, verbose=0)
print('Train: %.3f, Test: %.3f' % (train_acc, test_acc))
```

```python
# predict probabilities for test set
yhat_probs = rnn.predict(x_test_data, verbose=0)
```

```python
# predict crisp classes for test set
#ROC AUC requires the predicted class probabilities (yhat_probs) as an argument instead of the predicted classes (yhat_classes).
yhat_classes = np.where(yhat_probs > 0.19, 1,0)
```

```python
import numpy as np
import pandas as pd
raw_alarm = pd.read_csv('/content/billa_netact_alarms.csv')
raw_alarm.head(3)
```

```python
raw_alarm[['Last Detected Date','Last Detected Time','Time Zone']] = raw_alarm['Last Time Detected'].str.split(" ", n=2, expand = True)
raw_alarm.drop(columns = ['Last Time Detected','First Time Detected'], inplace = True)
raw_alarm.head(3)
```

```python
raw_alarm['Last Detected'] = raw_alarm['Last Detected Date']+' '+raw_alarm['Last Detected Time']
raw_alarm['Last Detected'] = pd.to_datetime(raw_alarm['Last Detected'])
raw_alarm.head(2)
```

```python
bb_raw_alarm = raw_alarm['Object Name'].str.split(",",n=2,expand = True)
bb_raw_alarm.head()
```

```python
raw_alarm['unit'] = bb_raw_alarm[0]
```

```python
alarm_data = pd.DataFrame(raw_alarm[['Last Detected','Severity','Probable Cause','Alarm Name','Alarm Type','unit']])
alarm_data.head(3)
```

```python
alarm_data.sort_values(by=['Last Detected'],inplace=True)

alarm_data.head(15)
```

```python
alarm_data.shape
```

```python
#Preparing the target column(Labeling our data as attach failed or not)
#For all the attach failure related alarms appeared y=1 else y=0
#Reading the attach alarms
attach_alarms = pd.read_csv('/content/Attach_alarms.csv')
print('The shape of attach_alarms is: ', attach_alarms)
print(attach_alarms)
```

```python
attach_alarm_list = list(attach_alarms.columns)
#for i in range(alarm_data.shape[0]):
  #for j in attach_alarm_list:
 # if alarm_data['Alarm Name'][i] in attach_alarm_list:
  #  alarm_data['y_target'][i] == 1.0
  #else:
   # alarm_data['y_target'][i] == 0.0
  
alarm_data['Attach Fail'] = [1 if x in attach_alarm_list else 0 for x in alarm_data['Alarm Name']]
print(alarm_data.head(4))
```

```python
alarm_data.head()
```

```python
alarm_data.shape
```

```python
#Encoding the Categorical Datas
df_categorical = pd.DataFrame(alarm_data[['Severity','Probable Cause','Alarm Name','Alarm Type','unit']])
df_categorical.head()
```

```python
df_categorical_encoded = pd.get_dummies(df_categorical, drop_first=True)
df_categorical_encoded.head()
```

```python
df_categorical_encoded.shape
```

```python
#final_dataset = pd.DataFrame(alarm_data['Last Detected'])
final_dataset = pd.concat([df_categorical_encoded,alarm_data['Attach Fail']], axis=1, ignore_index=False)
final_dataset.head(6)
```

```python
final_dataset.shape
```

```python
#Defining the train and test data
X = final_dataset.drop(columns = 'Attach Fail', axis =1)
Y = final_dataset['Attach Fail']
print('The shape of X features is: ',X.shape)
print('The shape of Y output is: ', Y.shape)
```

```python
X.index
```

```python
X.columns
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test,Y_train, Y_test = train_test_split(X,Y, test_size=0.15)

print('The shape of X features is: ',X.shape)
print('The shape of Y output is: ', Y.shape)
print('The shape of X_train features is: ',X_train.shape)
print('The shape of X_test  is: ', X_test.shape)
print('The shape of Y_train is: ',Y_train.shape)
print('The shape of Y_test is: ', Y_test.shape)
```

```python
print('The type of X features is: ',type(X))
print('The type of Y output is: ', type(Y))
print('The type of X_train features is: ',type(X_train))
print('The type of X_test  is: ', type(X_test))
print('The type of Y_train is: ',type(Y_train))
print('The type of Y_test is: ', type(Y_test))
```

```python
#converting our train data into numpy array as NN takes only numpy in tensorflow
x_training_data = np.array(X_train)

y_training_data = np.array(Y_train)
print(x_training_data.shape)

print(y_training_data.shape)
```

```python
x_training_data = np.expand_dims(x_training_data, axis = -1)
print(x_training_data.shape)
```

```python
x_training_data = np.reshape(x_training_data, (x_training_data.shape[0],x_training_data.shape[1],1))
print(x_training_data.shape)
```

```python
print(x_training_data.shape[0])
print(x_training_data.shape[1])
```

```python
from tensorflow.keras.models import Sequential

from tensorflow.keras.layers import Dense

from tensorflow.keras.layers import LSTM

from tensorflow.keras.layers import Dropout
```

```python
from keras.api._v2.keras import activations
rnn = Sequential()
rnn.add(LSTM(units = 45, return_sequences = True, input_shape = (x_training_data.shape[1],1)))
#input should be 3D with shape [batch, timesteps, feature]
rnn.add(Dropout(0.2))

rnn.add(LSTM(units = 45, return_sequences = True))

rnn.add(Dropout(0.2))

rnn.add(LSTM(units = 45, return_sequences = True))

rnn.add(Dropout(0.2))

rnn.add(LSTM(units = 45))

rnn.add(Dropout(0.2))
rnn.add(Dense(units = 1, activation = 'sigmoid'))
```

```python
rnn.summary()
```

```python
import tensorflow as tf
rnn.compile(optimizer = 'adam', loss = tf.keras.losses.BinaryCrossentropy(from_logits = True))
```

```python
rnn.fit(x_training_data, y_training_data, epochs = 100, batch_size = 32)
```

```python
#converting our test data into numpy array as NN takes only numpy in tensorflow
x_test_data = np.array(X_test)

y_test_data = np.array(Y_test)
print(x_test_data.shape)

print(y_test_data.shape)
```

```python
x_test_data = np.expand_dims(x_test_data, axis = -1)
print(x_test_data.shape)
```

```python
predictions = rnn.predict(x_test_data)
```
